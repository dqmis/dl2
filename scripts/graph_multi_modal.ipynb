```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

direct_sudoku_fill_in = [28.50, 25.51, 23.50]
direct_sudoku_validity = [47.25, 48.03, 49.50]
direct_graph_fill_in = [36.65, 29.50, 26.00]
direct_graph_validity = [54.75, 54.00, 52.51]
direct_set_validity = [58.00, 55.51, 53.01]

asp_sudoku_fill_in = [67.50, 22.00, 29.64]
asp_sudoku_validity = [92.08, 93.00, 87.75]
asp_graph_fill_in = [65.01, 38.21, 28.01]
asp_graph_validity = [94.50, 90.50, 86.25]
asp_set_validity = [67.00, 48.51, 54.51]




categories = ['Gemini Pro', 'Gemini Flash', 'GPT-4']
x = np.arange(len(categories))

bar_width = 0.15

direct_values = [direct_sudoku_fill_in, direct_sudoku_validity, direct_graph_fill_in, direct_graph_validity, direct_set_validity]
asp_values = [asp_sudoku_fill_in, asp_sudoku_validity, asp_graph_fill_in, asp_graph_validity, asp_set_validity]


for name, (sudoku_fill_in, sudoku_validity, graph_fill_in, graph_validity, set_validity) in [("Direct", direct_values), ("ASP", asp_values)]:
    plt.figure()

    fig, ax = plt.subplots()

    ax.bar(x - 2*bar_width, sudoku_fill_in, width=bar_width, label='Sudoku fill-in')
    ax.bar(x - bar_width, sudoku_validity, width=bar_width, label='Sudoku validity')
    ax.bar(x, graph_fill_in, width=bar_width, label='Graph fill-in')
    ax.bar(x + bar_width, graph_validity, width=bar_width, label='Graph validity')
    ax.bar(x + 2*bar_width, set_validity, width=bar_width, label='SET validity')

    ax.set_xlabel('model')
    ax.set_ylabel('accuracy (%)')
    ax.set_xticks(x)
    ax.set_xticklabels(categories, rotation=12)
    ax.legend()

    ax.legend(loc='upper left')

    ax.set_ylim(0, 100)
    ax.set_title(f'{name} prompting', fontsize=20)
    plt.tight_layout()
    plt.savefig(name + '.png')
```
